```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
df = pd.read_csv('all_data.csv')
print(df.head())
```

```python
df.Country.unique()
```

```python
df.Year.unique()
```

```python
df.rename(columns={'Life expectancy at birth (years)': 'LEABY'}, inplace=True)
print(df.head())
```

```python
sns.barplot(x='Country', y='GDP', data=df,)
country_list = df.Country.unique()
plt.show()
plt.savefig("Country Bar.png")
```

```python
sns.barplot(x='Country', y='LEABY', data=df)
plt.show()
plt.savefig("Country LEABY.png")
```

```python
sns.violinplot(y='Country', x='LEABY', data=df)
plt.show()
plt.savefig("Violin LEABY.png")
```

```python
f, ax = plt.subplots(figsize=(10, 15))
ax=sns.barplot(x='Country', y='GDP', hue='Year', data=df)
ax.set_xticklabels(df.Country.unique(), rotation=90)
plt.ylabel('GDP in Trillions of U.S. Dollars')
plt.show()
plt.savefig("GDPin Trillions.png")
```

```python
f, ax = plt.subplots(figsize=(10, 15))
ax=sns.barplot(x='Country', y='LEABY', hue='Year', data=df, palette='colorblind')
ax.set_xticklabels(df.Country.unique(), rotation=90)
plt.ylabel('Life expectancy at birth in years')
plt.show()
plt.savefig("LEABY per year.png")
```

```python
g = sns.FacetGrid(df, col='Year', hue='Country', col_wrap=4, height=2)
g = (g.map(plt.scatter, 'GDP', 'LEABY', edgecolor="w").add_legend())
plt.show()
plt.savefig("scatter LEABY by year country.png")
```

```python
g = sns.FacetGrid(df, col='Country', col_wrap=3, height=4)
g = (g.map(plt.plot, 'Year', 'LEABY').add_legend())
plt.show()
plt.savefig("line LEABY by year.png")
```

```python
g = sns.FacetGrid(df, col='Country', col_wrap=3, height=4)
g = (g.map(plt.plot, 'Year', 'GDP').add_legend())
plt.show()
plt.savefig("line GDP by year.png")
```
